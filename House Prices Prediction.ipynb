# Outlier Detection and Correction

```python
for i in rawDf.columns:
    if i!='Sale_Price':
        if rawDf[i].dtypes!='O':
            Q1 = np.quantile(rawDf[i],0.25)
            Q3 = np.quantile(rawDf[i],0.75)
            IQR = Q3-Q1
            LowerBound = Q1-1.5*IQR
            UpperBound = Q3+1.5*IQR
            rawDf[i]=np.where(rawDf[i]<LowerBound, LowerBound, rawDf[i])
            rawDf[i]=np.where(rawDf[i]>UpperBound, UpperBound, rawDf[i])
```

# Model Building

```python
from sklearn.linear_model import LinearRegression
```

```python
M1 = LinearRegression().fit(trainX,trainY)
```

Prediction

```python
test_pred = M1.predict(testX)
```

Evaluation

```python
RMSE_Error = np.sqrt(MS_Var)
```
